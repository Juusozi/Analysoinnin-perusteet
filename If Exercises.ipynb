```python
#Comparing two numbers

number1 = 6
number2 = 5

if number1 == number2:
    print("number1 and number2 are equal")

if number1 > number2:
    print("number1 is greater than number2")  

if number1 >= number2:
    print("number1 is greater or equal to number2")

if number1 != number2:
    print("number1 is NOT equal to number2")          
```

```python
#Comparing three numbers

number1 = 3
number2 = 2
number3 = 3

if number1 == number2 == number3:
    print("All numbers are equal")

if number1 == number2 or number2 == number3:
    print("Equal pair found") 

if number1 > number2 and number1 > number3:
    print("number1 is the greatest")

if number1 > number2:
    print("number1 is greater than number2")
else:
    if number2 > number3:
        print("number2 is greater than number3")            

if number1 == number2:
    print("number1 and number2 are equal")
elif number1 == number3:
    print("number1 and number3 are equal")    
```

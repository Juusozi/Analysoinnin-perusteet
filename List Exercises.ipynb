```python
#Exercise 2

import random

thrownDiceNumbers = []

for i in range(5):
    thrownDiceNumbers.append(random.randint(1, 6))

print("Thrown dice numbers:", thrownDiceNumbers)

print("sum:", sum(thrownDiceNumbers))

print("Highest value:", max(thrownDiceNumbers))

```
